```python
try:
    from bs4 import BeautifulSoup
    import requests
    import pandas as pd
    import http.client

except Exception as e:
    print(e)

else:
    print('import successfull')

import warnings
warnings.filterwarnings('ignore')
```

```python
url='https://www.goodreads.com/list/show/1.Best_Books_Ever?page='
books_name=[]
author=[]
ratings=[]
score_col=[]
vote_col=[]
for page in range(1,11):
    curl=url+str(page)
    payload = {'api_key': '65bbea6e986f43204c82fb7394a12ba8', 'url': curl}
    html= requests.get('http://api.scraperapi.com', params=payload)
    soup=BeautifulSoup(html.text,'lxml')
    box=soup.find('table',class_='tableList js-dataTooltip')

    books=box.find_all('a',class_='bookTitle')
    for book in books:
        books_name.append(book.text)

    auths=box.find_all('a',class_='authorName')
    for auth in auths:
        author.append(auth.text)

    stars=box.find_all('span',class_='minirating')
    for star in stars:
        ratings.append(star.text)

    scores=box.find_all('a',onclick="Lightbox.showBoxByID('score_explanation', 300); return false;")
    for score in scores:
        score_col.append(score.text)

    votes=box.find_all('span',class_='greyText')
    for vote in votes:
        check=vote.find_next_sibling('a')
        if check:
            vote_col.append(check.text)

    print(page,end=' ')
```

```python
print(len(books_name))
print(len(author))
print(len(ratings))
print(len(score_col))
print(len(vote_col))
```

```python
df=pd.DataFrame({'Book_Name':books_name,'Author':author,'Rating':ratings,'Score':score_col,'Votes':vote_col})
df.head()
```
